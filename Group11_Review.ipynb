#**Model Building**#

```python
dataset_features =fd.iloc[:, 0:24].values
dataset_labels = fd.iloc[:, 24].values
```

```python
from sklearn.model_selection import train_test_split
train_features, test_features, train_labels, test_labels = train_test_split(dataset_features, dataset_labels, test_size=0.3, random_state=21) 
```

```python
#model building # regression
from sklearn.linear_model import LogisticRegression
lr_object = LogisticRegression()
```

```python
lr_object.fit(train_features, train_labels)
```

```python
predicted_labels = lr_object.predict(test_features)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(classification_report(test_labels, predicted_labels))
```

```python
print(confusion_matrix(test_labels, predicted_labels))
```

```python
print(accuracy_score(test_labels, predicted_labels)) 
```

```python
#random forest
from sklearn.ensemble import RandomForestClassifier as rfc
rfc_object = rfc(n_estimators=200, random_state=0) 
rfc_object.fit(train_features, train_labels) 
```

```python
predicted_labels = rfc_object.predict(test_features)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(test_labels, predicted_labels))
```

```python
print(confusion_matrix(test_labels, predicted_labels))
```

```python
print(accuracy_score(test_labels, predicted_labels))
```

```python
#decision trees
from sklearn import tree
model = tree.DecisionTreeClassifier(criterion="entropy")
```

```python
model.fit(train_features, train_labels)
```

```python
predicted_labels = model.predict(test_features)
```
